### Importando o pandas

```python
import pandas as pd
```

### Criando um dataframe a partir de um dicionário

```python
# dataframe = pd.DataFrame()
venda = {'data': ['15/02/2021', '16/02/2021'],
         'valor': [500, 300],
         'produto': ['feijao', 'arroz'],
         'qtde': [50, 70],
        }

vendas_df = pd.DataFrame(venda) # é bom notificar no nome da variavel que é um df(dataframe),
                                # tbm pode colocar tabela (tabela_vendas , por exemplo)
```

### Importando arquivos e bases de dados

```python
vendas_df = pd.read_excel('Vendas.xlsx')
display(vendas_df)
```

### Pegar 1 coluna (e os pd.Series)

```python
produtos = vendas_df['Produto'] #quando vc pega só uma coluna, não é um dataframe, e sim uma serie.

produtos2 = vendas_df[['Produto', 'ID Loja']] # quando vc quer pegar mais de 1 coluna, usamos uma lista.
                                              # e sim, usamos 2 []

display(produtos2)
```

### .loc, um método muito importante
    - Pegar 1 linha
    - Pegar linhas de acordo com alguma condição
    - Pegar linhas e colunas específicas
    - Pegar 1 valor específico

```python
# 1 # pegar uma linha
display(vendas_df.loc[1:5])

# 2 # pegar linhas que correspondem a uma condição
vendas_norteshopping_df = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping']
display(vendas_norteshopping_df)

# 3 # pegar várias linhas e colunas usando o loc # 		 loc(linha,coluna)
vendas_norteshopping_df = vendas_df.loc[vendas_df['ID Loja'] == 'Norte Shopping',['ID Loja', 'Produto', 'Quantidade'] ]
display(vendas_norteshopping_df)

# 4 # pegar um valor específico
print(vendas_df.loc[1,'Produto'])
```

### Adicionar 1 coluna

```python
# 1. A partir de uma coluna que existe - se vc tentar "editar" uma coluna que não existe,
# o pandas cria essa coluna 
vendas_df['Comissão'] = vendas_df['Valor Final'] * 0.05

# 2. Criar uma coluna padrão
# o : serve para falar "todas as linhas (ou colunas)" para o pandas
vendas_df.loc[: , 'Imposto'] = 0
```

### Adicionar 1 linha
    - Linhas de um complemento da base de dados

```python
#vendas_dez_df = pd.read_excel('Vendas - Dez.xlsx')
#display(vendas_dez_df)

#vendas_df = vendas_df.append(vendas_dez_df)
display(vendas_df)
```

### Excluir linhas e colunas

```python
vendas_df = vendas_df.drop('Imposto', axis=1)
# axis = eixo ---- 1 = coluna, 0 = linha
display(vendas_df)
```

### Valores Vazios
    - Deletar linhas/colunas vazias
    - Deletar linhas que possuem valores vazios
    - Preencher valores vazios (média e último valor)

```python
# Deletar linhas/colunas vazias COMPLETAMENTE vazias
vendas_df = vendas_df.dropna(how='all') # o parametro força a excluir APENAS se todos os dados forem nulos

# deletar linhas que possuem pelo menos 1 valor vazio
vendas_df = vendas_df.dropna()

# preencher valores vazios
# 1. com a média da coluna
vendas_df['Comissão'] = vendas_df['Comissão'].fillna(vendas_df['Comissão'].mean())

# 2. com o último valor acima
vendas_df = vendas_df.ffill()
```

### Calcular Indicadores
    - Groupby
    - Value Counts

```python
# Value Counts
#transações_por_loja = vendas_df['ID Loja'].value_counts()
#display(transações_por_loja)

# Groupby
faturamento_por_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').sum()
display(faturamento_por_produto)
```

### Mesclar 2 dataframes (Procurar informações de um dataframe em outro)

```python
gerentes_df = pd.read_excel('Gerentes.xlsx')
#display(gerentes_df)
vendas_df = vendas_df.merge(gerentes_df)
display(vendas_df)
```
